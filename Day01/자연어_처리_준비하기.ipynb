# Series
- 1차원 배열의 값에 각 값(values)에 대응되는 인덱스(index)를 부여할 수 있는 구조

````python
sr = pd.Series([17000, 18000, 1000, 5000],
               index=["피자", "치킨", "콜라", "맥주"])
print('시리즈 출력 :')
print('-'*15)
print(sr)
````

````python
print('시리즈의 값 : {}'.format(sr.values))
print('시리즈의 인덱스 : {}'.format(sr.index))
````

# DataFrame
- 2차원 리스트를 매개변수로 전달
- 행방향 인덱스(index)와 열방향 인덱스(column)
- column, index, values 구성


````python
values = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
index = ['one', 'two', 'three']
columns = ['A', 'B', 'C']

df = pd.DataFrame(values, index=index, columns=columns)

print('데이터프레임 출력 :')
print('-'*18)
print(df)
````

````python
print('데이터프레임의 인덱스 : {}'.format(df.index))
print('데이터프레임의 열이름: {}'.format(df.columns))
print('데이터프레임의 값 :')
print('-'*18)
print(df.values)
````

# Numpy 배열
- 개념: 축의 개수(ndim), 크기(shape)
- 배열의 크기를 정확히 숙지하는 것이 딥러닝에서 중요
```
np.zeros((ndim, shape))
-> 배열의 모든 원소에 0 삽입

np.ones((ndim, shape))
-> 배열의 모든 원소에 1을 삽입

np.full((ndim, shape), NUMBER)
-> 배열에 NUMBER 값 삽입

np.eye(대각선의 개수)
-> 대각선의 개수: 3일 때, 3*3 크기의 배열 생성
-> 대각선만 1, 나머지 0인 대각행렬 생성

np.random.random((ndim, shape))
-> ndim*shape 크기의 배열 생성 및 임의의 값 삽입

np.arange(NUMBER)
-> 0부터 NUMBER-1 까지의 값을 가지는 1차원 배열 생성

np.arange(START, END, STEP)
-> START부터 END-1까지 STEP만큼 증가하는 배열 생성

np.reshape()
-> 내부 데이터는 변경하지 않으면서 배열의 구조 변경
-> 예시(아래 코드 참조)

```

````python
reshape_mat = np.array(np.arange(30)).reshape((6,5))
print(reshape_mat)
````

# Machine Learning Workflow
머신_러닝_워크플로우.png
1. 수집
- 자연어 데이터를 말뭉치 또는 코퍼스(corpus)라고 부름

2. 점검 및 탐색
- 데이터의 구조, 노이즈 데이터, 머신 러닝 적용을 위해서 정제하는 방법 파악
- 탐색적 데이터 분석 단계
- 시각화와 간단한 통계 테스트 진행

3. 전처리 및 정제
- 자연어 처리의 경우 토큰화, 정제, 정규화, 불용어 제거
- 빠르고 정확한 데이터 전처리를 하기 위해서는 다양한 라이브러리에 대한 지식 필요

4. 모델링 및 훈련
- 머신 러닝 알고리즘을 통해 학습(훈련)

- 현업의 경우 검증용 데이터는 거의 필수

5. 평가
