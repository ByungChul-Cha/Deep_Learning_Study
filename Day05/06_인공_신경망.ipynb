# **06-01. 머신 러닝 기본 개념 정리하기**

# 머신 러닝 모델의 평가
- 검증용 데이터:
  - 모델의 성능을 조정하기 위한 용도
  - 과적합이 되고 있는지 판단
  - 하이퍼파리미터의 조정을 위한 용도
- 하이퍼파리미터(초매개변수): 값에 따라서 모델의 성능에 영향을 주는 매개변수들
- 파라미터(매개변수): 학습을 통해 바뀌어져가는 변수(가중치, 편향)
- k-폴드 교차 검증: 검증 데이터와 테스트 데이터를 나눌 만큼 데이터가 충분하지 않을 때 사용

# Classification & Regression
- 회귀 문제 학습: Linear Regression, Logistic Regression
- 분류 문제 학습: Logistic Regression
  - Binary Classification
  - Mulit-Class Classification
  - Multi-label Classification으로 나뉨

# Supervised Learning(지도 학습)과 Unsupervised Learning(비지도 학습)
- 지도 학습
  - 레이블(정답)과 함께 학습하는 것
- 비지도 학습
  - 목적 데이터(레이블)이 없는 학습 방법
  - 군집(clustering)이나 차원 축소와 같은 학습 방법
- 강화 학습
  - 어떤 환경 내에서 정의된 에이전트가 현재의 상태를 인식,
  - 선택 가능한 행동들 중 보상을 최대화하는 행동 혹은 행동 순서를 선택하는 방법

# Sample & Feature
- Sample(샘플): 하나의 데이터, 행렬에서는 하나의 행
  - DB에서는 레코드라고 부르는 단위
- Feature(특성): 종속 변수 y를 예측하기 위한 각각의 독립 변수 x

# Confusion Matrix(혼동 행렬)
- Accuracy(정확도): 맞춘 문제 수를 전체 문제 수로 나눈 값
- 혼동 행렬:
  - 정확도는 맞춘 결과와 틀린 결과에 대한 세부적인 내용을 안 알려줌
    - TP(True Positive): 양성이라고 대답, 양성이 정답
    - TN(True Negative): 음성이라고 대답, 음성이 정답
    - FP(False Positive): 양성이라고 대답, 음성이 정답
    - FN(False Negative): 음성이라고 대답, 양성이 정답
  - Precision(정밀도)
    - 양성이라고 대답한 전체 케이스에 대한 TP의 비율
    - TP / (TP + FP)
  - Recall(재현률)
    - 실제값이 양성인 데이터의 전체 개수에 대해서 TP의 비율
    - Tp / (TP + FN)

# Overfitting & Underfitting
- Overfitting(과적합): 훈련 데이터를 과하게 학습한 경우
  - 훈련 데이터에 대해서는 정확도가 높음
  - 테스트 데이터의 성능이 낮아지기 전에 훈련을 멈추는 것이 바람직
  - 방지: Dropout(드롭아웃), Early Stopping(조기 종료)
- Underfitting(과소적합): 성능이 올라갈 여지가 있음에도 훈련을 덜 한 상태
  - 훈련 자체가 부족한 상태
  - 훈련 데이터에 대해서도 보통 정확도가 낮음

# 비선형 활성화 함수
- 입력을 받아 수학적 변환을 수행하고 출력을 생성하는 함수
- 인공 신경망의 은닉층에서 활성화 함수로 ReLU 함수를 사용
  - Sigmoid 함수는 지양

1. **활성화 함수의 특징 - Nonlinear function(비선형 함수)**
  - 선형 함수가 아닌 비선형 함수여야 함
  - 선형 함수로는 은닉층을 여러번 추가하더라도 1회 추가한 것과 차이를 줄 수 없음
  - 선형 함수를 사용한 층을 표현하기 위한 방법
    - 선형층(linear layer)
    - 투사층(projection layer)
  - 활성화 함수를 사용하는 층: 비선형 층(nonlinear layer)
2. **Sigmoid 함수와 기울기 소실**
  - 기울기 소실:
    - 역전파 과정에서 0에 가까운 아주 작은 기울기가 곱해지게 되면
    - 앞단에는 잘 전달 되지 않음
3. **Hyperbolic tangent function(하이퍼볼릭 탄젠트 함수)**
  - 입력값을 -1과 1 사이의 값으로 변환
  - 시그모이드 함수보다는 기울기 소실 증상이 적은 편
  - 은닉층에서 시그모이드 함수보다는 많이 사용
4. **ReLU 함수**
  - 수식: f(x) = max(0, x)
    - 음수를 입력하면 0을 출력
      - 문제점: 입력값이 음수면 기울기도 0 -> dying ReLU 문제 발생
    - 양수를 입력하면 입력값 그대로 반환
    - 단순 임계값이므로 연산 속도가 빠름
    - 특정 양수값에 수렴하지 않음 -> 깊은 신경망에서 잘 작동
5. **Leaky ReLU**
  - dying ReLU 문제 보완
  - 수식: f(x) = max(ax, x)
    - a는 hyperparameter, 일반적으로 0.01의 값을 가짐
6. **Softmax function**
  - 분류 문제를 Logistic Regression, Softmax Regression를 출력층에 적용하여 사용
7. **출력층의 활성화 함수와 오차 함수의 관계**

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Sigmoid function
def sigmoid(x):
  return 1/(1+np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.plot([0,0],[1.0, 0.0], ':')
plt.title('Sigmoid Function')
plt.show()
````

````python
# Hyperbolic tanget function
x = np.arange(-5.0, 5.0, 0.1)
y = np.tanh(x)

plt.plot(x, y)
plt.plot([0,0],[1.0,-1.0], ':')
plt.axhline(y=0, color='orange', linestyle='--')
plt.title('Hyperbolic Tanh Function')
plt.show()
````

````python
# ReLU function
def relu(x):
  return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.plot([0,0],[5.0, 0.0], ":")
plt.title("ReLU Function")
plt.show()
````

````python
# Leaky ReLU function
a = 0.1
def leaky_relu(x):
  return np.maximum(a*x, x)

x = np.arange(-5.0, 5.0, 0.1)
y = leaky_relu(x)

plt.plot(x, y)
plt.plot([0,0],[5.0,0.0], ':')
plt.title('Leaky ReLU Function')
plt.show()
````

````python
# Softmax function
x = np.arange(-5.0, 5.0, 0.1)
y = np.exp(x) / np.sum(np.exp(x))

plt.plot(x, y)
plt.title('Softmax Function')
plt.show()
````

# **06-02. Perceptron(퍼셉트론)**

# Perceptron(퍼셉트론)
- 초기 형태의 인공 신경망
- 다수의 입력으로부터 하나의 결과를 내보내는 알고리즘
- 입력값에는 각각의 가중치가 존재하는데, 가중치의 값이 클수록 해당 입력 값이 중요
- 계단 함수에 사용된 임계치 값을 보통 θ로 표현
- 활성화 함수: 뉴런에서 출력값을 변경시키는 함수

# Single-Layer Perceptron(단층 퍼셉트론)
- 값을 보내는 단계(입력층)와 값을 받아서 출력하는 단계(출력층)로 이루어짐
- 직선 하나로 두 영역을 나눌 수 있는 문제에 대해서 구현 가능
  - AND, OR, NAND를 구현
  - 한계:
    - XOR 게이트를 구현하지 못함 -> 비선형 영역으로 분리할 수 있음
    - 선형 영역에 대해서만 분리 가능

````python
def AND_gate(x1, x2):
  w1 = 0.5
  w2 = 0.5
  b = -0.7
  result = x1*w1 + x2*w2 + b
  if result <= 0:
    return 0
  else:
    return 1

def NAND_gate(x1, x2):
  w1 = -0.5
  w2 = -0.5
  b = 0.7
  result = x1*w1 + x2*w2 + b
  if result <= 0:
    return 0
  else:
    return 1

def OR_gate(x1, x2):
    w1=0.6
    w2=0.6
    b=-0.5
    result = x1*w1 + x2*w2 + b
    if result <= 0:
        return 0
    else:
        return 1
````

# MultiLayer Perceptron, MLP(다층 퍼셉트론)
- 단층 퍼셉트론 + 중간에 층을 더 추가(은닉층)
- 심층 신경망(Deep Neural Network, DNN): 은닉층이 2개 이상인 신경망
- 학습: 기계가 가중치를 스스로 찾아내도록 자동화시키는 것
- 딥 러닝(Deep Learning): 인공 신경망이 심층 신경망일 경

````python
# AND, NAND, OR 게이트 조합해서 XOR 게이트 구현
def XOR_gate(x1, x2):
  s1 = NAND_gate(x1, x2)
  s2 = OR_gate(x1, x2)
  result = AND_gate(s1, s2)
  if result == 0 :
    return 0
  else:
    return 1
````

# **06-03. XOR 문제 구현하기- 단층 퍼셉트론과 다층 퍼셉트론**

# pytorch로 단층 퍼셉트론 구현하기

````python
import torch
import torch.nn as nn

# GPU 연산이 가능할 경우 GPU 연산으로 수행
device = 'cuda' if torch.cuda.is_available() else 'cpu'
torch.manual_seed(777)
if device == 'cuda':
  torch.cuda.manual_seed_all(777)
````

````python
X = torch.FloatTensor([[0,0], [0,1], [1,0], [1,1]]).to(device)
Y = torch.FloatTensor([[0], [1], [1], [0]]).to(device)
````

````python
# 입력 2개 -> 출력 1개로 변환
linear = nn.Linear(2, 1, bias=True)
sigmoid = nn.Sigmoid()
model = nn.Sequential(linear, sigmoid).to(device)
````

````python
# 비용 함수와 옵티마이저 정의
# BCELoss()는 이진 분류에서 사용하는 크로스엔트로피 함수
criterion = nn.BCELoss().to(device)
optimizer = torch.optim.SGD(model.parameters(), lr=1)
````

````python
for step in range(10001):
  optimizer.zero_grad()
  hypothesis = model(X)

  cost = criterion(hypothesis, Y)
  cost.backward()
  optimizer.step()

  if step % 1000 == 0:
    print('{:5d}: {:.6f}'.format(step, cost.item()))
````

````python
with torch.no_grad():
  hypothesis = model(X)
  predicted = (hypothesis > 0.5).float()
  accuracy = (predicted==Y).float().mean()
  print('모델의 출력값(Hypothesis): ', hypothesis.detach().cpu().numpy())
  print('모델의 예측값(Predicted): ', predicted.detach().cpu().numpy())
  print('실제값(Y): ', Y.cpu().numpy())
  print('정확도(Accuracy): ', accuracy.item())
````

````python
with torch.no_grad():
  hypothesis = model(X)
  predicted  = (hypothesis > 0.5).float()
  accuracy = (predicted == Y).float().mean()
  print('모델의 출력값(Hypothesis): ', hypothesis.detach().cpu().numpy())
  print('모델의 예측값(Predicted): ', predicted.detach().cpu().numpy())
  print('실제값(Y): ', Y.cpu().numpy())
  print('정확도(Accuracy): ', accuracy.item())
````

# **06-06. 다층 퍼셉트론으로 MNIST 분류하기**

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import fetch_openml
````

````python
# cache=True를 통해 다운로드한 데이터를 로컬 디스크에 저장하여
# 다음 호출 때 빠르게 로드
mnist = fetch_openml('mnist_784', version=1, cache=True, as_frame=False)
````

````python
mnist.target[0]
````

````python
# 레이블 데이터 타입을 정수형으로 변환
mnist.target = mnist.target.astype(np.int8)
````

````python
mnist.target[0]
````

````python
X = mnist.data / 255
y = mnist.target
````

````python
plt.imshow(X[0].reshape(28,28), cmap='gray')
print("이 이미지 데이터의 레이블은 {:.0f}이다".format(y[0]))
````

# 훈련 데이터와 테스트 데이터의 분리

````python
import torch
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/7,
                                                    random_state = 0)
````

````python
# 텐서로 변환
X_train = torch.Tensor(X_train)
X_test = torch.Tensor(X_test)
y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)

# TensorDataset 객체 생성
ds_train = TensorDataset(X_train, y_train)
ds_test = TensorDataset(X_test, y_test)

# DataLoader 객체 생성
loader_train = DataLoader(ds_train, batch_size=64, shuffle=True)
loader_test = DataLoader(ds_test, batch_size=64, shuffle=False)
````

# 다층 퍼셉트론

````python
from torch import nn
from torch import optim
````

````python
model = nn.Sequential()
model.add_module('fc1', nn.Linear(28*28*1, 100))
model.add_module('relu1', nn.ReLU())
model.add_module('fc2', nn.Linear(100, 100))
model.add_module('relu2', nn.ReLU())
model.add_module('fc3', nn.Linear(100, 10))

print(model)
````

````python
# 오차함수 선택
loss_fn = nn.CrossEntropyLoss()

# 최적화 기법 선택
optimizer = optim.Adam(model.parameters(), lr=0.01)
````

````python
epochs = 3
for epoch in range(epochs):
  for data, targets in loader_train:
    optimizer.zero_grad()
    y_pred = model(data)
    loss = loss_fn(y_pred, targets)
    loss.backward()
    optimizer.step()

  print("Epoch {:4d}/{} Cost: {:.6f}".format(epoch + 1, epochs, loss.item()))
````

````python
# 신경망을 추론 모드로 전환
model.eval()
correct = 0

# 데이터로더에서 미니 배치를 하나씩 꺼내 추론을 수행
with torch.no_grad():
  for data, targets in loader_test:
    outputs = model(data)

    # 추론 계산
    # 확률이 가장 높은 레이블이 무엇인지 계산
    # torch.max는 반환값이 2개 -> 최대값, 인덱스
    _, predicted = torch.max(outputs.data, 1)
    correct += predicted.eq(targets.data.view_as(predicted)).sum()

# 정확도 출력
data_num = len(loader_test.dataset)
print("Accuracy: {}/{} ({:.0f}%)".format(correct, data_num, 100.* correct / data_num))
````

# **06-07. 과적합을 막는 방법들**

1. **데이터의 양을 늘리기**
  - 양이 적을 경우
    - 데이터 증식 또는 증강: 의도적으로 기존의 데이터를 조금씩 변형하고 추가
2. **모델의 복잡도 줄이기**
  - 레이어의 개수 줄이기
  - 수용력: 모델에 있는 매개변수들의 수
3. **가중치 규제(Regularization) 적용하기**
  - L1 규제: 가중치들의 절대값 합계를 비용 함수에 추가
    - L1 노름이라고도 함
    - 어떤 특성들이 모델에 영향을 주고 있는지를 정확히 판단
  - L2 규제: 모든 가중치들의 제곱합을 비용 함수에 추가
    - L2 노름이라고도 함
    - 가중치 감쇠(weight decay)라고도 함
    - 더 잘 동작
4. **Dropout(드롭아웃)**
  - 학습 과정에서 신경망의 일부를 사용하지 않는 방법
  - 매 뉴런마다 해당 확률로 결
  - 신경망 학습 시에만 사용
  - 예측 시에는 사용하지 않는 것이 일반적
