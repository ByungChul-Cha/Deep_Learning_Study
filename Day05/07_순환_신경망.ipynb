# **07-01. Recurrent Neural Network(RNN, 순환신경망)**

# Recurrent Neural Network, RNN
- 입력과 출력을 시퀀스 단위로 처리하는 모델
- 가장 기본적인 시퀀스 모델
- Feed Forwar Neural Network:
  - 은닉층에서 활성화 함수를 지난 값이 오직 출력층 방향으로 향하는 것
- 출력층 방향으로 보내면서, 다시 은닉층 노드의 다음 계산의 입력으로 보내는 특징
- Cell:
  - 은닉층에서 활성화 함수를 통해 결과를 내보내는 역할
  - 메모리 셀 또는 RNN 셀이라고도 표현
- Hidden State:
  - 메모리 셀이 출력층 방향으로 보내는 값
  - 다음 시점의 자신에게 보내는 값
- 입력 벡터, 출력 벡터, 은닉층에서는 은닉 상태라는 표현 사용
```
RNN의 다양한 형태
1. one-to-many(일 대 다)
  - Image Captioning 작업에 사용
    - 하나의 이미지 입력에 대해서 사진의 제목 출력
2. many-to-one(다 대 일)
  - Sentiment Classification(감성 분류)
    - 입력 문서가 긍정적인지 부정적인지를 판별
  - Spam Detection(스팸 메일 분류)
    - 메일이 정상 메일인지 스팸 메일인지 판별
3. many-to-many(다 대 다)
  - Chatbot
    - 입력 문장으로부터 대답 문장 출력
  - 번역기
  - 개체명 인식
  - 품사 태깅
```

# 파이썬으로 RNN 구현하기

````python
import numpy as np

# 시점의 수, NLP에서는 보통 문장의 길이가 됨
timesteps = 10
# 입력의 차원, NLP에서는 보통 단어 벡터의 차원이 됨
input_size = 4
# 은닉 상태의 크기, 메모리 셀의 용량
hidden_size = 8

# 입력에 해당되는 2D 텐서
inputs = np.random.random((timesteps, input_size))

# 은닉 상태의 크기 hidden_size로 은닉 상태 만듦
hidden_state_t = np.zeros((hidden_size,))
````

````python
print(hidden_state_t)
````

````python
Wx = np.random.random((hidden_size, input_size))
Wh = np.random.random((hidden_size, hidden_size))
b = np.random.random((hidden_size, ))
````

````python
print(np.shape(Wx))
print(np.shape(Wh))
print(np.shape(b))
````

````python
total_hidden_states = []

# 메모리 셀 동작
# 각 시점에 따라서 입력값이 입력됨
for input_t in inputs:
  output_t = np.tanh(np.dot(Wx, input_t) + np.dot(Wh, hidden_state_t) + b)

  # 각 시점의 은닉 상태의 값을 계속해서 축적
  total_hidden_states.append(list(output_t))
  print(np.shape(total_hidden_states))

  hidden_state_t = output_t

total_hidden_states = np.stack(total_hidden_states, axis=0)

print(total_hidden_states)
````

# 파이토치의 nn.RNN()

# Bidirectional Recurrent Neural Network(양방향 순환 신경망)
- 시점 t에서의 출력값을 예측할 때 이전 시점의 데이터뿐만 아니라

- 하나의 출력값을 예측하기 위해 기본적으로 2개의 메모리 셀 사용

  - 뒤 시점의 은닉 상태

````python
cell = nn.RNN(input_size = 5, hidden_size = 8, num_layers = 2,
              bidirectional=True, batch_first=True)
````

````python
outputs, _status = cell(inputs)
````

# **07-02. LSTM & GRU**

# 바닐라 RNN의 한계
- 비교적 짧은 시퀀스에 대해서만 효과
- 장기 의존성 문제:
  - 앞부분에 나온 정보에 대한 영향력이 의미가 없어지는(중요성 소실) 문제

# LSTM(Long Short-Term Memory)
- 메모리 셀에 다음을 추가
  - 입력 게이트
  - 망각 게이트(삭제 게이트)
  - 출력 게이트
  - 셀 상태(cell state)
- 긴 시퀀스의 입력을 처리하는데 탁월한 성능
- 입력 게이트:
  - 현재 정보를 기억하기 위한 게이트
  - 나온 2개의 값을 통해 기억할 정보의 양을 정함
  - 현재 시점의 입력을 얼마나 반영할지 결정
- 삭제 게이트:
  - 기억을 삭제하기 위한 게이트
  - 시그모이드 함수를 통해 나온 값이 삭제 과정을 거친 정보의 양
  - 0에 가까울수록 정보가 많이 삭제
  - 이전 시점의 입력을 얼마나 반영할지 결정
- 셀 상태(장기 상태)
  - 입력 게이트에서 구한 2개의 값에 원소별 곱 진행
  - 입력 게이트에서 선택된 기억을 삭제 게이트의 결과값과 더함
- 출력 게이트와 은닉 상태(단기 상태)
  - 현재 시점 t의 은닉 상태를 결정하는 일
  - 값이 걸러지는 효과 발생
  - 출력층으로도 향함

# Gated Recurrent Unit(GRU)
- 업데이트 게이트와 리셋 게이트 존재
- 데이터 양이 적으면 GRU
- 데이터 양이 많으면 LSTM

# **07-03. 문자 단위 RNN(Char RNN)**

# 문자 단위 RNN
- 문자 단위 RNN: RNN의 입출력의 단위가 문자 레벨
- 다대다 구조로 구현

````python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
````

````python
input_str = "apple"
label_str = "pple!"
# 중복을 제거한 문자들의 집합
char_vocab = sorted(list(set(input_str+label_str)))
vocab_size = len(char_vocab)
print("문자 집합의 크기 : {}".format(vocab_size))
````

````python
input_size = vocab_size
hidden_size = 5
output_size = 5
learning_rate = 0.1
````

````python
char_to_index = dict((c, i) for i,c in enumerate(char_vocab))
print(char_to_index)
````

````python
index_to_char = {}
for key, value in char_to_index.items():
  index_to_char[value] = key
print(index_to_char)
````

````python
x_data = [char_to_index[c] for c in input_str]
y_data = [char_to_index[c] for c in label_str]
print(x_data)
print(y_data)
````

````python
# nn.RNN()은 기본적으로 3D 텐서를 입력받음
# 배치 차원 추가
# unsqueeze(0)을 통해 해결 가능
x_data = [x_data]
y_data = [y_data]
print(x_data)
print(y_data)
````

````python
# 원-핫 벡터로 바꿈
x_one_hot = [np.eye(vocab_size)[x] for x in x_data]
print(x_one_hot)
````

````python
X = torch.FloatTensor(x_one_hot)
Y = torch.LongTensor(y_data)
````

````python
print("훈련 데이터의 크기 : {}".format(X.shape))
print("레이블의 크기 : {}".format(Y.shape))
````

````python
# RNN 모델 구현
class Net(nn.Module):
  def __init__(self, input_size, hidden_size, output_size):
    super(Net, self).__init__()
    # RNN 셀 구현
    self.rnn = nn.RNN(input_size, hidden_size, batch_first=True)
    # 출력층 구현
    self.fc = nn.Linear(hidden_size, output_size, bias=True)

  def forward(self, x):
    x, _status = self.rnn(x)
    x = self.fc(x)
    return x
````

````python
net = Net(input_size, hidden_size, output_size)
````

````python
outputs = net(X)
# 배치 차원, 시점, 출력의 크기
print(outputs.shape)
````

````python
# 2D 텐서로 변환
print(outputs.view(-1, input_size).shape)
````

````python
print(Y.shape)
print(Y.view(-1).shape)
````

````python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(net.parameters(), learning_rate)
````

````python
for i in range(100):
  optimizer.zero_grad()
  outputs = net(X)

  # view를 하는 이유는 Batch 차원 제거를 위해
  loss = criterion(outputs.view(-1, input_size), Y.view(-1))
  loss.backward()
  optimizer.step()

  # 실제 어떻게 예측했는지 확인
  result = outputs.data.numpy().argmax(axis=2)
  result_str = ''.join([index_to_char[c] for c in np.squeeze(result)])
  print("{:2d} loss: {} prediction {} trueY: {} prediction str: {}"
  .format(i, loss.item(), result, y_data, result_str))
````

# 더 많은 데이터로 학습한 문자 단위 RNN

````python
sentence = ("if you want to build a ship, don't drum up people together to "
            "collect wood and don't assign them tasks and work, but rather "
            "teach them to long for the endless immensity of the sea.")
````

````python
char_set = list(set(sentence))
char_dic = {c: i for i, c in enumerate(char_set)}
````

````python
print(char_dic)
````

````python
dic_size = len(char_dic)
print("문자 집합의 크기: {}".format(dic_size))
````

````python
hidden_size = dic_size
sequence_length = 10
learning_rate = 0.1
````

````python
x_data = []
y_data = []
for i in range(0, len(sentence) - sequence_length):
  x_str = sentence[i:i + sequence_length]
  y_str = sentence[i + 1:i + sequence_length + 1]
  print(i, x_str, '->', y_str)

  x_data.append([char_dic[c] for c in x_str])
  y_data.append([char_dic[c] for c in y_str])
````

````python
print(x_data[0])
print(y_data[0])
````

````python
x_one_hot = [np.eye(dic_size)[x] for x in x_data]
X = torch.FloatTensor(x_one_hot)
y = torch.LongTensor(y_data)
````

````python
print("훈련 데이터의 크기:",X.shape)
print("레이블의 크기:", y.shape)
````

````python
class Net(nn.Module):
  def __init__(self, input_dim, hidden_dim, layers):
    super(Net, self).__init__()
    self.rnn = nn.RNN(input_dim, hidden_dim, num_layers=1, batch_first=True)
    self.fc = nn.Linear(hidden_dim, hidden_dim, bias=True)

  def forward(self, x):
    x, _status = self.rnn(x)
    x = self.fc(x)
    return x
````

````python
net = Net(dic_size, hidden_size, 2)
````

````python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(net.parameters(), learning_rate)
````

````python
outputs = net(X)
print(outputs.shape)
````

````python
print(outputs.view(-1, dic_size).shape)
````

````python
print(y.shape)
````

````python
print(y.view(-1).shape)
````

````python
for i in range(100):
  optimizer.zero_grad()
  outputs = net(X)
  loss = criterion(outputs.view(-1, dic_size), y.view(-1))
  loss.backward()
  optimizer.step()

  # results의 크기는 (170, 10)
  results = outputs.argmax(dim = 2)
  predict_str = ""
  for j, result in enumerate(results):
    if j == 0:
      predict_str += ''.join([char_set[t] for t in result])
    else:
      predict_str += char_set[result[-1]]
  print(predict_str)
````
