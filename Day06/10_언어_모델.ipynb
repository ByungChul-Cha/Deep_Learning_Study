# **10-01. 언어 모델(Language Model)이란?**

# Language Model(언어 모델)
- 단어 시퀀스에 확률을 할당하는 모델
  - 통계를 이용한 방법
  - 인공 신경망을 이용한 방법(더 좋은 성능)
- 이전 단어들이 주어졌을 때 다음 단어를 예측
- Language Modeling(언어 모델링):
  - 주어진 단어들로부터 아직 모르는 단어를 예측하는 작업
  - 언어 모델이 이전 단어들로부터 다음 단어를 예측하는 일


# 단어 시퀀스의 확률 할당
1. 기계 번역(Machine Translation)
2. 오타 교정(Spell Correction)
3. 음성 인식(Speech Recognition)

# **10-02. 통계적 언어 모델과 N-gram 언어 모델**
- 통계적 언어 모델: Statistical Language Model, SLM

# 카운트 기반의 접근
- SLM은 이전 단어로부터 다음 단어에 대한 확률을 카운트에 기반하여 계산
- 한계:
  - Sparsity Problem(희소 문제):
    - 충분한 데이터를 관측하지 못하여 언어를 정확히 모델링하지 못하는 문제

# N-gram 언어 모델
- 일부 단어만 고려하는 접근 방법
- N-gram에서 n이 몇 개의 단어를 보느냐 결정
- 앞 단어 중 n개만 포함해서 카운트하여 근사하는 것
- n: 연속적인 단어 나열
  - n이 1일 때: unigram
  - n이 2일 때: bigram
  - n이 3일 때: trigram
  - n이 4 이상일 때: 숫자를 그대로 사용
- n=4일 때 , n-1에 해당되는 앞의 3개의 단어만 고려

# **10-03. Perplexity, PPL**

# Evaluation metric: PPL
- PPL: 언어 모델을 평가하기 위한 평가 지표
  - **낮을수록 언어 모델의 성능이 좋은 것**
  - 문장의 길이로 정규화된 문장 확률의 역수

#  Branching factor(분기 계수)
- PPL: 선택할 수 있는 가능한 경우의 수를 의미하는 분기계수
  - 언어 모델이 특정 시점에서 평균적으로 몇 개의 선택지를 가지고 고민하고 있는지
- 주의사항:
  - 테스트 데이터에 의존
  - 2개 이상의 언어 모델을 비교할 때
    - 정량적으로 양이 많아야함
    - 도메인에 알맞은 동일한 테스트 데이터를 사용
